```
# importing libraries
import re
import pandas as pd
import requests
from urllib.request import urlopen as ureq
from requests import get
from bs4 import BeautifulSoup as BS
import time
import numpy as np
from urllib.parse import urlparse
from tqdm import tqdm_notebook as tqdm
```

```
links_howstat = pd.DataFrame(columns=['Year', 'Link'])
r = requests.get('http://howstat.com/cricket/Statistics/WorldCup/MatchListMenu.asp')
soup = BS(r.content,"html5lib")
rows = soup.find("table", attrs={'class':'TableLined'}).find("tbody").find_all("tr")
for row in rows:
  year = row.find("td").text.strip()
  li = row.find("td").find_all("a", href=True)
  for link in li:
    href = 'http://howstat.com/cricket/Statistics/WorldCup/'+link['href']
    links_howstat = links_howstat.append(pd.Series([year, href], index=links_howstat.columns ), ignore_index=True)
links_howstat
```

```
match_info = pd.DataFrame(columns=['Match ID', 'Match Date', 'Year' ,'Team 1', 'Team 2', 'Winner', 'Result', 'Groud', 'Link'])

for link in tqdm(links_howstat['Link'], desc='Progress'):
  r = requests.get(link)
  soup = BS(r.content,"html5lib")
  rows = soup.find("table", attrs={'class':'TableLined'}).find("tbody").find_all("tr")
  for row in rows:
    ground = row.find("td").find_next("td").find_next("td").text.strip()
    if ground != 'Ground / Location':
      li = row.find("td").find_next("td").find_all("a", href=True)
      for link in li:
        href = 'http://howstat.com/cricket/Statistics/WorldCup/'+link['href']
        m_id = re.search('\d+',str(href)).group(0)
        match = link.text
        start1 = match.find( ':' )
        start2 = match.find( 'v' )
        team1 = match[start1+1:start2].strip()
        team2 = match[start2+1:].strip()
        bracket = team2.find( '(' )
        if bracket != -1:
          team2 = team2[:bracket].strip()
        m_date = row.find("td").text.strip()
        year = m_date.split('/')[2].strip()
        result = row.find("td").find_next("td").find_next("td").find_next("td").text.strip()
        if team1 in result:
          winner = team1
        else:
          winner = team2
      match_info = match_info.append(pd.Series([str(m_id), m_date, year, team1, team2, winner, result, ground, href], index=match_info.columns ), ignore_index=True)
match_info
```

```
match_info.to_csv('World Cup All Matches Info.csv', index=False)
```

```
# Batting Stats
batting_stats = pd.DataFrame(columns=['Match ID','Year','Innings','Country', 'Player','Wicket','R','BF','4s','6s'])

for i in tqdm(range(0, len(match_info['Link'])), desc='Progress'):
  if match_info['Result'][i] != 'Match abandoned' and match_info['Result'][i] != 'No result' and 'default' not in str(match_info['Result'][i]):
    r = requests.get(match_info['Link'][i])
    soup = BS(r.content,"html5lib")
    innings = ['Innings 1','Innings 2']
    count = 0


    # Innings 1
    table = soup.findAll('table')[7]
    rows = table.find_all('tr')[0:13]
    country = table.find_all('tr')[0].find("td").text.strip()
    for row in rows:
      if row.find("td").find_next("td").text.strip() != 'R' and row.find("td").text.strip() != 'Extras':
        link = row.find('a', href=True)
        player_url = 'http://www.howstat.com/cricket/Statistics/WorldCup/'+link['href']
        r_player = requests.get(player_url)
        soup_player = BS(r_player.content,"html5lib")
        table_player = soup_player.findAll('table')[6]
        player = table_player.find("tbody").find_all('tr')[2].find("td").find_next("td").text.strip()
        wicket = row.find("td").find_next("td").text.strip()
        wicket = wicket.replace('*','')
        wicket = wicket.replace('†','')    
        runs = row.find("td").find_next("td").find_next("td").text.strip()
        bf = row.find("td").find_next("td").find_next("td").find_next("td").text.strip()
        fours = row.find("td").find_next("td").find_next("td").find_next("td").find_next("td").text.strip()
        sixes = row.find("td").find_next("td").find_next("td").find_next("td").find_next("td").find_next("td").text.strip()
        batting_stats = batting_stats.append(pd.Series([match_info['Match ID'][i], match_info['Year'][i], innings[0], country, player, wicket, runs, bf, fours, sixes], index=batting_stats.columns ), ignore_index=True)
        
    
    
    # Innings 2
    table = soup.findAll('table')[7]
    rows = table.find_all('tr')
    for row in rows:
      if row.find('td').find('table') != None:
        no_of_rows = len(row.find('td').find('table').find('tbody').find_all('tr'))
        row_count = count + no_of_rows + 2
        break
      count = count+1
    rows2 = table.find_all('tr')[row_count+1:row_count+13]
    country2 = rows2[0].find("td").text.strip()
    start = country2.find( '(' )
    if start != -1:
      country2 = country2[0:start].strip()
    for row in rows2:
      if row.find("td").find_next("td").text.strip() != 'R' and row.find("td").text.strip() != 'Extras':
        link2 = row.find('a', href=True)
        player_url2 = 'http://www.howstat.com/cricket/Statistics/WorldCup/'+link2['href']
        r_player2 = requests.get(player_url2)
        soup_player2 = BS(r_player2.content,"html5lib")
        table_player2 = soup_player2.findAll('table')[6]
        player2 = table_player2.find("tbody").find_all('tr')[2].find("td").find_next("td").text.strip()
        wicket2 = row.find("td").find_next("td").text.strip()
        wicket2 = wicket2.replace('*','')
        wicket2 = wicket2.replace('†','')      
        runs2 = row.find("td").find_next("td").find_next("td").text.strip()
        bf2 = row.find("td").find_next("td").find_next("td").find_next("td").text.strip()
        fours2 = row.find("td").find_next("td").find_next("td").find_next("td").find_next("td").text.strip()
        sixes2 = row.find("td").find_next("td").find_next("td").find_next("td").find_next("td").find_next("td").text.strip()
        batting_stats = batting_stats.append(pd.Series([match_info['Match ID'][i], match_info['Year'][i], innings[1], country2, player2, wicket2, runs2, bf2, fours2, sixes2], index=batting_stats.columns ), ignore_index=True)
```
