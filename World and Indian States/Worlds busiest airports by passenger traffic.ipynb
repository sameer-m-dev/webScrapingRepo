```
# importing libraries
import re
import pandas as pd
import requests
from urllib.request import urlopen as ureq
from requests import get
from bs4 import BeautifulSoup as BS
import time
import numpy as np
from urllib.parse import urlparse
from tqdm import tqdm_notebook as tqdm
```

```
airport_data = pd.DataFrame(columns=['Airport', 'Country', 'Code'])
airport_data = airport_data.fillna(0)
airport_data
```

```
# links_howstat = pd.DataFrame(columns=['Year', 'Link'])

i = 2018
Flag = True
while Flag:
  if i > 2009:
    r = requests.get('https://en.wikipedia.org/wiki/List_of_busiest_airports_by_passenger_traffic')
    soup = BS(r.content,"html5lib")
    tables = soup.find_all("table")[:9]
    for table in tables:
    #   print(table)
      rows = table.find("tbody").find_all("tr")
      for row in rows[1:]:
        name = row.find('td').find_next('td').text.strip()
        country = row.find('td').find_next('td').find_next('td').find_next('td').text.strip()
        code = row.find('td').find_next('td').find_next('td').find_next('td').find_next('td').text.strip()
        traffic = row.find('td').find_next('td').find_next('td').find_next('td').find_next('td').find_next('td').text.strip()
        if code not in set(airport_data['Code']):
          row_num = len(airport_data['Airport'])
          airport_data.loc[row_num,'Airport'] = name
          airport_data.loc[row_num,'Country'] = country
          airport_data.loc[row_num,'Code'] = code
        else:
          row_num = airport_data.index[airport_data['Code'] == code][0]
        airport_data.loc[row_num,'%s' % i] = traffic
      i = i-1
  elif i < 2010 and i > 2004: 
    r = requests.get('https://en.wikipedia.org/wiki/List_of_busiest_airports_by_passenger_traffic_(2000%E2%80%932009)')
    soup = BS(r.content,"html5lib")
    tables = soup.find_all("table")[:5]
    i = 2009
    # print(tables)
    for table in tables:
    #   print(table)
      rows = table.find("tbody").find_all("tr")
      for row in rows[1:]:
        name = row.find('td').find_next('td').text.strip()
        country = row.find('td').find_next('td').find_next('td').find_next('td').text.strip()
        code = row.find('td').find_next('td').find_next('td').find_next('td').find_next('td').text.strip()
        traffic = row.find('td').find_next('td').find_next('td').find_next('td').find_next('td').find_next('td').text.strip()
        if code not in set(airport_data['Code']):
          row_num = len(airport_data['Airport'])
          airport_data.loc[row_num,'Airport'] = name
          airport_data.loc[row_num,'Country'] = country
          airport_data.loc[row_num,'Code'] = code

        else:
          row_num = airport_data.index[airport_data['Code'] == code][0]
        airport_data.loc[row_num,'%s' % i] = traffic
      i = i-1
  elif i <= 2004:
    r = requests.get('https://en.wikipedia.org/wiki/List_of_busiest_airports_by_passenger_traffic_(2000%E2%80%932009)')
    soup = BS(r.content,"html5lib")
    tables = soup.find_all("table")[5:9]
    i = 2004
    # print(tables)
    for table in tables:
    #   print(table)
      rows = table.find("tbody").find_all("tr")
      for row in rows[1:]:
        name = row.find('td').find_next('td').text.strip()
        country = row.find('td').find_next('td').find_next('td').text.strip()
        code = row.find('td').find_next('td').find_next('td').find_next('td').text.strip()
        traffic = row.find('td').find_next('td').find_next('td').find_next('td').find_next('td').text.strip()
        if code not in set(airport_data['Code']):
          row_num = len(airport_data['Airport'])
          airport_data.loc[row_num,'Airport'] = name
          airport_data.loc[row_num,'Country'] = country
          airport_data.loc[row_num,'Code'] = code

        else:
          row_num = airport_data.index[airport_data['Code'] == code][0]
        airport_data.loc[row_num,'%s' % i] = traffic

      i = i-1
    Flag = False
```
