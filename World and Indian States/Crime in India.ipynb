```
# importing libraries
import re
import pandas as pd
import requests
from urllib.request import urlopen as ureq
from requests import get
from bs4 import BeautifulSoup as BS
import time
import numpy as np
from urllib.parse import urlparse
from tqdm import tqdm_notebook as tqdm
```

```
total_suicides = pd.read_csv('Total Suicides in Indian States and UTs (2001-2015).csv')
```

```
states_group_by_year = total_suicides.groupby('YEAR')
zeros_arr = np.zeros(shape=(len(set(total_suicides['STATE/UT'])), 16), dtype=int)
sut_crimes_per_year = pd.DataFrame(zeros_arr, columns=['STATE/UT', '2001',
 '2002',
 '2003',
 '2004',
 '2005',
 '2006',
 '2007',
 '2008',
 '2009',
 '2010',
 '2011',
 '2012',
 '2013',
 '2014'])
count = 0
for year in tqdm(set(total_suicides['YEAR']), desc='Progress'):
  all_states = states_group_by_year.get_group(year).reset_index().drop('index', axis=1)
  
  for i in range(0, len(all_states['STATE/UT'])):
    if all_states['STATE/UT'][i] not in set(sut_crimes_per_year['STATE/UT']):
      row_num = count
      count = count+1
      sut_crimes_per_year.loc[row_num,'STATE/UT'] = all_states['STATE/UT'][i]
    else:
      row_num = sut_crimes_per_year.index[sut_crimes_per_year['STATE/UT'] == all_states['STATE/UT'][i]][0]
    sut_crimes_per_year.loc[row_num,'%s' % str(year)] += int(all_states['TOTAL'][i])

```
